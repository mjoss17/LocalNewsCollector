# Tranform Data Dump from Wikipedia

```python
# Import dependencies
import pandas as pd
import re
```

## Import & Inspect Data

```python
# Import data
extract_dump_df = pd.read_csv('dump.csv', index_col=('Region', 'Subregion', 'State/Territory', 'Newspaper'), encoding='utf-8')
extract_dump_df.head()
```

```python
# Drop empty columns (may want to include a thresh down the line)
extract_dump_df = extract_dump_df.dropna(how='all', axis=1)
extract_dump_df.head()
```

```python
# Get overview of data
extract_dump_df.info()
```

```python
# View column value spread
extract_dump_df.count()
```

## Bin Catagorical Data

```python
# Inspect catagorical columns
for col in ['Type', 'Format', 'Language', 'Political alignment']:
    print("\n---- %s ---" % col)
    print(extract_dump_df[col].value_counts())
```

```python
# Group labels for Type column
new_labels = {
    'Daily newspaper': ['Daily', 'Daily Newspaper', 'Daily newspaper', 'Daily ', 'daily'],
    'Weekly newspaper': ['Weekly', 'Weekly ', 'Weekly Newspaper', 'Weekly newspaper', 'weekly', 'weekly newspaper'],
    'Biweekly newspaper': ['Biweekly', 'Biweekly newspaper', 'Twice-weekly Newspaper', 'Twice weekly Newspaper', 
                            'Twice-weekly', 'bi weekly ', 'Bi-weekly', 'Alternative weekly'],
    'Triweekly newspaper': ['Thrice Weekly ', 'Triweekly newspaper', 'Local thrice-weekly Paper', 'Three times per week '],
    'Monthly newspaper': ['monthly bulletin', 'Monthly newspaper', 'Monthly ', 'Monthly Newpaper'],
    'Online newspaper': ['External site', 'News Agency, Online', 'News site']
}
transform = {key: orig_key for orig_key, orig_val in new_labels.items() for key in orig_val}
extract_dump_df['Type'] = extract_dump_df['Type'].apply(lambda x: transform.get(x, 'Other'))
extract_dump_df['Type'].value_counts()
```

```python
# Group labels for Format column
new_labels = {
    'Broadsheet': ['Broadsheet', 'broadsheet'],
    'Tabloid': ['Tabloid', 'tabloid', '[tabloid]'],
    'Compact': ['Compact', 'compact'],
    'Print': ['Print', 'A3', 'Magazine'],
    'Online': ['Online', 'Online newspaper', 'Digital-only']
}
transform = {key: orig_key for orig_key, orig_val in new_labels.items() for key in orig_val}
extract_dump_df['Format'] = extract_dump_df['Format'].apply(lambda x: transform.get(x, 'Other'))
extract_dump_df['Format'].value_counts()
```

```python
# Group labels for Political alignment column
new_labels = {
    'Centre-right': ['Centre-right', 'Centre-Right', 'centre-right'],
    'Centre': ['Centre', 'centre', 'Centrist', 'Centrism'],
    'Liberal': ['Liberal', 'Liberalism'],
    'Conservative': ['Conservative', 'Conservatism'],
    'Other': ['At present: ', '<>']
}
transform = {key: orig_key for orig_key, orig_val in new_labels.items() for key in orig_val}
extract_dump_df['Political alignment'] = extract_dump_df['Political alignment'].apply(lambda x: transform.get(x, x))
extract_dump_df['Political alignment'].value_counts()
```

## Convert Datatypes

```python
# View current datatypes
extract_dump_df.dtypes
```

```python
# Extract year from Founded string
def extractYear(date_string):
    if pd.notnull(date_string):
        if match := re.search(r'(\d{4})', date_string):
            return int(match[0])
    return date_string # Might want to have a defalut val here, or more ways to extract date. Want all ints/NULL

extract_dump_df['Founded'] = extract_dump_df['Founded'].apply(extractYear)
extract_dump_df['Founded'].describe()
```
